### Feature Engineering/Extraction

- `gp`, `pts`

A player's scoring ability might be a better predictor of their career longevity than the number of games they've played. While experience is valuable, a player's impact on the game, as measured by points scored, could be a stronger indicator of their potential for a long career.  While the total number of games played might not be a strong predictor alone, combining it with points scored could create a valuable metric. This new feature would give us a better understanding of a player's overall performance.
***
- `gp`, `min`

Combining points scored with minutes played (`min * gp`) could create a valuable efficiency metric. This new feature, along with total points scored, might be better predictors of career length than the individual components alone.
***
- `fgm`, `fga`, `fg`
- `3p_made`, `3pa`, `3p`
- `ftm`, `fta`, `ft`

Percentages (fg%, 3p%, ft%) are more informative than raw numbers for made or attempted. These percentages provide a better measure of a player's shooting efficiency by considering both successful and unsuccessful attempts, and they also allow for better comparisons to other players.
***
- `reb` = `oreb` + `dreb`

The total rebound count (`reb`) is sufficient and should be used over the individual offensive and defensive rebound counts (`oreb` and `dreb`). This provides a comprehensive measure of rebounding performance, which is all that is needed for assessing career durations.
***
- `ast`, `stl`, `blk`, `tov`

Assists, steals, blocks, and turnovers are key indicators of a player's overall performance and could significantly impact their career longevity.
***
##### Features to keep
At this stage, these features are deemed most relevant for modeling:

`gp`, `min`, `pts`, `fg`, `3p`, `ft`, `reb`, `ast`, `stl`, `blk`, `tov`, `target_5yrs`


```python
# Save selections to new df
df2 = df1[["gp", "min", "pts", "fg", "3p", "ft", "reb", "ast", "stl", "blk", "tov", "target_5yrs"]]
df2.head()
```

All features are numerical, therefore no additional encoding of categorical features is needed.

```python
# Extracting relevant features
df3 = df2.copy()
```

```python
# Create new column: `total_points` 
# Calculate total points earned by multiplying the number of games played 
# by the average number of points earned per game
df3['total_points'] = df3['gp'] * df3['pts']
```

```python
# Create new column:`efficiency`
# Calculate efficiency by dividing the total points earned 
# by the total number of minutes played, which yields points per minute.
# (Note that `min` represents avg. minutes per game.)
df3["efficiency"] = df3["total_points"] / (df3["min"] * df3["gp"])
```

```python
df3.head()
```

### Data Cleaning

#### Drop remaining unused columns

```python
# Remove columns no longer needed
df3 = df3.drop(columns=['gp', 'pts', 'min'])
df3.head()
```

## Part 2: Model Building and Evaluation

#### Prepare data and fit the model

```python
# Isolate the features (X)
y = df3['target_5yrs']

# Isolate target variable (y)
X = df3.drop('target_5yrs', axis = 1)
```

##### Create training and holdout sets

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.25, random_state=0)
```

##### Fit the model to the training data

```python
# Instantiate
nb = naive_bayes.GaussianNB()

# Fit the model to the training data
nb.fit(X_train, y_train)
```

#### Evaluate Model

Assess model performance on training data to determine next steps.

```python
# Predict on test data
y_pred = nb.predict(X_test)
```

```python
# Get scoring metrics
print('accuracy score:'), print(metrics.accuracy_score(y_test, y_pred))
print('precision score:'), print(metrics.precision_score(y_test, y_pred))
print('recall score:'), print(metrics.recall_score(y_test, y_pred))
print('f1 score:'), print(metrics.f1_score(y_test, y_pred))
```

Plot a confusion matrix of the model's predictions on the test data.

```python
# Generate array of values for confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred)

# Plot confusion matrix
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb.classes_)
disp.plot(cmap = "Oranges", values_format='');
```

- The diagonal from the top left to the bottom right of the confusion matrix highlights correctly classified instances. The proportion of correct predictions to total instances is the accuracy of the model (0.6896).  Accuracy alone is not the best score to rely on.
  
- The model's high precision score (0.8406) is due to a significantly larger number of true positives compared to false positives.

- The model's recall score is lower (0.5856) due to a smaller difference between false negatives and true positives.

- The F1 score of 0.6905 balances precision and recall and give a combines assessment.  This score suggests reasonable predictive power in this model.

- The model is more likely to misclassify players with long careers as having short careers than the other way around. This indicates the higher rate of Type II errors.

Having an F1 score of 0.6905, it is very good at correctly identifying positive cases (high precision of 84.06%) but less effective at finding all positive cases (low recall of 58.56%).

The model tends to misclassify players with long careers as having short careers almost 4x more often than the other way around, indicating a higher rate of Type II errors.

## Part 3: Conclusion/Summary

Features were used to help predict NBA career durations: field goals, three-point field goals, free throws, rebounds, assists, steals, blocks, turnovers, total points, and efficiency as points per minute.

The steps that were taken:
- Feature selection - `fg`, `3p`, `ft`, `reb`, `ast`, `stl`, `blk`, `tov`, `target_5yrs`
- Feature extraction:
  - Created `total points` to measure efficiency by points scored (number of games played * average number of points earned per game)
  - Created `efficiency` to measure points per minute by dividing the total points earned by the total number of minutes played (`total_points` / (`min` * `gp`) )

### **Conclusion**

The model created provides some value in predicting an NBA player's chances of playing for five years or more.  It is very good at correctly identifying positive cases but less effective at finding all positive cases.

Particularly, the model tends to misclassify players with long careers as having short careers almost 4x more often than the other way around.
